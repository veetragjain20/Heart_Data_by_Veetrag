```python
top_HeartDiseasePresence=df["HeartDiseasePresence"].value_counts().head(10)
print("top 10 HeartDiseasePresence")
print(top_HeartDiseasePresence)
plt.figure(figsize=(2,3))
sns.countplot(x="HeartDiseasePresence",data=df)
```

```python
top_Thalassemia=df["Thalassemia"].value_counts().head(10)
print("top 10 Thalassemia")
print(top_Thalassemia)
plt.figure(figsize=(4,1))
sns.countplot(x="Thalassemia",data=df)
```

```python
top_NumMajorVessels=df["NumMajorVessels"].value_counts().head(10)
print("NumMajorVessels")
print(top_NumMajorVessels)
plt.figure(figsize=(4,3))
sns.countplot(x="NumMajorVessels",data=df)
```
