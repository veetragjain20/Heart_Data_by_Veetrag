```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r"/content/heart_data_set.csv")
print(df)
print(df.head(10))
print("Dataset Shape:", df.shape)
```

```python
missing_values_count = df.isnull().sum()
print("missing values in each column:")
print(missing_values_count)
```

```python
print("Number of unique values in each column:")
for column in df.columns:
    print(f"{column}: {df[column].nunique()} unique values")
    # print(f"Unique values for {column}: {df[column].unique()}") # Uncomment this line if you want to see the actual unique values
```

```python
average_age=df['age'].mean()
print("Average Age:", average_age)
```

```python
min_age=df['age'].min()
print("Minimum Age:", min_age)
```

```python
max_age=df['age'].min()
print("Maximum Age:", max_age)
```

```python
heart_disease_count=df['HeartDiseasePresence'].value_counts()
print("Heart Disease Counts:")
print(heart_disease_count)
```

```python
import matplotlib.pyplot as plt
gender_counts=df['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel("Gender'(0=Female, 1=Male)")
plt.ylabel("Number of Patients")
plt.title("Number of Patients by Gender")
plt.show()
```

```python
import seaborn as sns
sns.histplot(df["age"], bins=20)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Patients")
plt.show()
```

```python
corr =df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap of Heart Disease Dataset")
plt.show()
```
